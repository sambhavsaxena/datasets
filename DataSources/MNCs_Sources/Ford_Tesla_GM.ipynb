```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start = datetime.datetime(2020, 1, 1)
end = datetime.datetime(2021, 6, 27)
```

```python
tesla = web.DataReader("TSLA", 'yahoo', start, end)
```

```python
ford = web.DataReader("F", 'yahoo', start, end)
```

```python
gm = web.DataReader("GM", 'yahoo', start, end)
```

```python
tesla.to_csv("Tesla_datasource.csv")
ford.to_csv("Ford_datasource.csv")
gm.to_csv("GM_datasource.csv")
```

```python
tesla['Open'].plot(label="Tesla", figsize = (16, 4.5))
gm['Open'].plot(label="GM")
ford['Open'].plot(label="Ford")
plt.title("Stock graph of shares of Tesla, Ford and GM")
plt.legend()
plt.ylabel('Stock Price')
```

```python
tesla['Volume'].plot(label = "Tesla", figsize = (16, 4.5))
gm['Volume'].plot(label = "GM")
ford['Volume'].plot(label = "Ford")
plt.legend()
plt.ylabel('Volume Traded')
```
