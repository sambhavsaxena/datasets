```python
import datetime
import pandas_datareader as web
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2021, 1, 1)
```

```python
end = datetime.datetime(2021, 6, 27)
```

```python
google = web.DataReader("GOOGL", 'yahoo', start, end)
```

```python
google.head()
```

```python
google['Close'].plot(label = 'GOOGL Close price')
google['Open'].plot(label = 'GOOGL Open price', figsize = (16, 4.5))
google['High'].plot(label = 'GOOGL High price')
google['Low'].plot(label = 'GOOGL Low price')
plt.legend()
plt.title("Google Stock Analysis")
plt.ylabel('Stock Price')
plt.show()
```
